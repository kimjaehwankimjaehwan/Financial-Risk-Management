```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn')

# Quandl API를 이용해 데이터 가져오기
import yfinance as yf
data = yf.download("OIL", start="1980-01-01", end="2020-01-01")
# data['Close'].plot(title="Oil Prices")


# 시각화
plt.figure(figsize=(10, 6))
plt.plot(data.Close)
plt.ylabel('$')
plt.xlabel('Date')
plt.show()
```
